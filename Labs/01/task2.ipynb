## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Preprocessing

```python
df = pd.read_csv("cancer_patient_dataset.csv")
df.head(5)
```

## Data Analysis

```python
#Part C:
df['Level'].unique()
```

```python
#Part D:
print("Total Columns:", len(df.columns))
print("\nTop 50 records:\n", df.head(50))
print("\nBottom 30 records:\n", df.tail(30))
print("\nDataset Info:\n")
df.info()
print("\nShape:", df.shape)
print("\nSize:", df.size)
```

```python
#Part E:
print("Number of Columns:", df.shape[1])
print("Total Samples:", df.shape[0])
```

```python
#Part F:
# Numerical
print(df['Level'].value_counts())

# Graphical
sns.countplot(x='Level', data=df)
plt.title("Distribution of Target Variable (Level)")
plt.show()
```

```python
#Part G:
new_df = df.loc[550:900, "Gender":"Coughing of Blood"]

new_df.to_csv("subset_loc.csv", index=False)

print("Mean:\n", new_df.mean())
print("Median:\n", new_df.median())
print("Mode:\n", new_df.mode().iloc[0])
print("Standard Deviation:\n", new_df.std())
print("Minimum:\n", new_df.min())
print("Maximum:\n", new_df.max())
```

```python
# Part H

last_20 = df.iloc[-20:, df.columns.get_loc("Shortness of Breath"):df.columns.get_loc("Frequent Cold")+1]
last_20.to_csv("last20_iloc.csv", index=False)

print("Mean:\n", last_20.mean())
print("Median:\n", last_20.median())
print("Mode:\n", last_20.mode().iloc[0])
print("Std Dev:\n", last_20.std())
print("Minimum:\n", last_20.min())
print("Maximum:\n", last_20.max())
```
