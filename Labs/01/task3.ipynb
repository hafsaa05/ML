```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("cancer_patient_dataset.csv")

print(df.head())
```

```python
# Part A - Graphical Visualization of "Age"
plt.figure(figsize=(8,5))
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df["Age"], color="lightgreen")
plt.title("Boxplot of Age")
plt.xlabel("Age")
plt.show()

# Part B
features = ["Age", "Smoking", "Obesity", "Alcohol use", "Shortness of Breath"]
corr_matrix = df[features].corr()

plt.figure(figsize=(7,5))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.3)
plt.title("Correlation Heatmap of Selected Features")
plt.show()

# Part C
plt.figure(figsize=(8,4))

# histogram for Age
plt.subplot(1,2,1)  # row=1 col=2 position=1
plt.hist(df["Age"], bins=15, color="skyblue", edgecolor="black")
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age")

# countplot for Frequent Cold (adjusted to remove warning)
plt.subplot(1,2,2)  # row=1 col=2 position=2
sns.countplot(x="Frequent Cold", data=df, color="lightpink")  # changed from palette
plt.title("Frequent Cold Distribution")

plt.tight_layout()
plt.show()

# Part D
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Obesity"], color="lightcoral")
plt.title("Outlier Detection in Obesity")
plt.show()
```
